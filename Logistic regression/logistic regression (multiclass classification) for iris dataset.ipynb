```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
myData = load_iris()
```

```python
myData.data
```

```python
myData.target
```

```python
myData.target_names
```

```python
myData.feature_names
```

```python
df = pd.DataFrame(myData.data,columns=myData.feature_names)
df
```

```python
df['target'] = myData.target
df
```

```python
df['flower_name'] = df['target'].apply(lambda x: myData.target_names[x])
df
```

```python
my_cols = ["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)","flower_name","target"]
df = df.reindex(columns=my_cols)
df
```

```python
a = df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]]
b = df[["target"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2)
```

```python
print(len(a_train))
print(len(a_test))
```

```python
from sklearn.linear_model import LogisticRegression
myModel = LogisticRegression()
```

```python
myModel.fit(a_train,b_train)
```

```python
myModel.score(a_test,b_test)
```

```python
b_test
```

```python
myModel.predict(a_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = myModel.predict(a_test)
cm = confusion_matrix(b_test,y_pred)
cm
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
```
