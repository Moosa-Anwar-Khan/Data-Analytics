```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.corr()
```

```python
df.info()
```

```python
ndf = df
ndf = ndf.drop(['PassengerId', 'Name', 'Parch','Ticket','Cabin','Embarked'], axis="columns")
ndf
```

```python
ndf.isnull().sum()
```

```python
pd.options.display.max_rows = 999
```

```python
ndf.isnull().sum()
```

```python
ndf['Age'].median()
```

```python
ndf.fillna(ndf['Age'].median(), inplace=True)
ndf
```

```python
myCols = ['Pclass','Sex','Age','SibSp','Fare','Survived']
ndf = ndf.reindex(columns=myCols)
ndf.head()
```

```python
target = ndf['Survived']
attributes = ndf.drop(['Survived'],axis="columns")
```

```python
attributes.head()
```

We will use "dummy variables" for "Sex" column. We could use LabelEncoder() but since our other columns consists of continuous data so thats why we will use "dummy variables"

```python
dummyVars = pd.get_dummies(attributes['Sex'])
dummyVars.head()
```

```python
attributes = pd.concat([attributes,dummyVars],axis="columns")
attributes.head()
```

```python
attributes = attributes.drop(['Sex'],axis="columns")
attributes.head()
```

```python
attributes.isna().any() # or we can use => attributes.isna().sum()  
```

```python
attributes.shape
```

```python
from sklearn.model_selection import train_test_split
a_train, a_test, b_train, b_test = train_test_split(attributes,target,test_size=0.2)
```

```python
print("Length of training data: ",len(a_train))
print("Length of testing data: ",len(a_test))
```

```python
a_train
```

```python
a_test
```

```python
from sklearn.naive_bayes import GaussianNB
myModel = GaussianNB()
```

```python
myModel.fit(a_train,b_train)
```

```python
myModel.score(a_test,b_test)
```

Lets check the actual result of first ten rows of testing dataset and then compare it with predicted result

```python
a_test.head()
```

```python
b_test.head()
```

```python
myModel.predict(a_test.head())
```

```python
myModel.predict_proba(a_test.head())
```

Lets use confusion matrix for checking the performance of our model

```python
from sklearn.metrics import confusion_matrix
y_pred = myModel.predict(a_test)
cm = confusion_matrix(b_test, y_pred)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True)
plt.xlabel("Actual", fontSize=15)
plt.ylabel("Predicted", fontSize=15)
```
