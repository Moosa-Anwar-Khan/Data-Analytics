```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("heartDisease.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df1 = df
```

```python
dum1 = pd.get_dummies(df1['ChestPainType'])
dum2 = pd.get_dummies(df1['RestingECG'])
dum3 = pd.get_dummies(df1['ExerciseAngina'])
dum4 = pd.get_dummies(df1['ST_Slope'])
dum5 = pd.get_dummies(df1['Sex'])
```

```python
pd.options.display.max_columns = 50
```

```python
conDf1 = pd.concat([df1,dum1,dum2,dum3,dum4,dum5],axis="columns")
conDf1.head()
```

```python
myDF1 = conDf1.drop(["Sex","ChestPainType","RestingECG","ExerciseAngina","ST_Slope"],axis="columns")
myDF1.head()
```

```python
df1['Oldpeak'].unique()
```

```python
myCols = ["Age","F","M","ATA","NAP","ASY","TA","RestingBP","Cholesterol","FastingBS",
"Normal","ST","LVH","MaxHR","N","Y","Oldpeak","Up","Flat","Down","HeartDisease"]
myDF1 = myDF1.reindex(columns=myCols)
myDF1.head()
```

```python
a_ran = myDF1.drop(["HeartDisease"],axis="columns") # ran for random 
b_ran = myDF1["HeartDisease"]
```

```python
a_ran.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

a_ran["Age"] = scaler.fit_transform(a_ran[["Age"]])
a_ran["RestingBP"] = scaler.fit_transform(a_ran[["RestingBP"]])
a_ran["Cholesterol"] = scaler.fit_transform(a_ran[["Cholesterol"]])
a_ran["MaxHR"] = scaler.fit_transform(a_ran[["MaxHR"]])
a_ran["Oldpeak"] = scaler.fit_transform(a_ran[["Oldpeak"]])
a_ran.head()
```

```python
from sklearn.model_selection import train_test_split
a_ran_train, a_ran_test, b_ran_train, b_ran_test = train_test_split(a_ran,b_ran,test_size=0.2)
```

```python
print("Length of training data set: ",len(a_ran_train))
print("Length of testing data set: ",len(a_ran_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(a_ran_train,b_ran_train)
```

```python
print("Accuracy: ",model.score(a_ran_test,b_ran_test)*100,"%")
```

```python
model.predict(a_ran_test)
```

Now we will apply PCA technique on our 'a_log' data:

```python
a_ran.head()
```

```python
a_ran.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
a_ran_pca = pca.fit_transform(a_ran)
a_ran_pca.shape
```

As we can in above two cells that number of columns decrease from 20 to 10 after applying PCA technique

```python
pca.explained_variance_ratio_
```

```python
aPca_ran_train, aPca_ran_test, b_ran_train, b_ran_test = train_test_split(a_ran_pca,b_ran,test_size=0.2)
```

```python
model2 = RandomForestClassifier()
model2.fit(aPca_ran_train, b_ran_train)
print("Accuracy: ",model2.score(aPca_ran_test,b_ran_test)*100,"%")
```

```python
b_ran_test
```

```python
model2.predict(aPca_ran_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model2.predict(aPca_ran_test)
cm = confusion_matrix(b_ran_test,y_pred)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
```
